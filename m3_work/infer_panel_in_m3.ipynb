```python
def get_age_pred(row):
    probs = [row['age_<=18'], row['age_19-29'], row['age_30-39'], row['age_>=40']]
    ind_max = probs.index(max(probs))
    if ind_max == 0:
        return '0-18'
    if ind_max == 1:
        return '19-29'
    if ind_max == 2:
        return '30-39'
    if ind_max == 3:
        return '40-999'
    
def get_gender_pred(row):
    if row['gender_male'] >= row['gender_female']:
        return 'Male'
    return 'Female'

result_concat['pred_gender'] = result_concat.apply(lambda x: get_gender_pred(x), axis=1)
result_concat['pred_age_range'] = result_concat.apply(lambda x: get_age_pred(x), axis=1)

result_concat['min_age'] = result_concat['pred_age_range'].apply(lambda x: int(x.split('-')[0]))
result_concat['max_age'] = result_concat['pred_age_range'].apply(lambda x: int(x.split('-')[1]))
```

## Compare to panel data

```python
panel = pd.read_csv('data\\panel_sample.csv')
panel.head(2)
```

```python
join = pd.merge(panel, result_concat, left_on='user_id', right_on='id')
```

```python
from datetime import date

def calculate_age(born, pit=date.today()):
    born = pd.to_datetime(born)
    return pit.year - born.year - ((pit.month, pit.day) < (born.month, born.day))

join['real_age'] = join.birth_date.apply(lambda x: calculate_age(x))
join.head(2)
```

```python
mismatch_gender = len(join[join['sex'] != join ['pred_gender']])
tot = len(join)
print("accuracy of gender predication is " + str((tot-mismatch_gender)/tot))
```

```python
mismatch_age = len(join[(join['min_age']  join ['real_age']) | (join['max_age'] < join ['real_age'])])
tot = len(join)
print("accuracy of age predication is " + str((tot-mismatch_age)/tot))
```

## Match against good annotations

```python
annot_join = pd.merge(annots, join, left_on='user_id', right_on='user_id')
```

```python
def get_m3_age_bucket(age):
    if age <= 18:
        return 'age_<=18'
    if age <= 29:
        return 'age_19-29'
    if age <= 39:
        return 'age_30-39'
    return 'age_>=40'

def check_for_annot_match(row):
    # first check which bin does each annotator belongs to and compare
    # second check if the annotators bin match real bin (dated to 2018 when the annotation took place)
    # if all agree return true
    annot1_age_grp = get_m3_age_bucket(row['Age (estimate a number)'])
    annot2_age_grp = get_m3_age_bucket(row['Age (estimate a number).1'])
    if annot1_age_grp != annot2_age_grp:
        return False
    age_2018 = calculate_age(row['birth_date'], pit=date(2018,8,1))
    age_2018_grp = get_m3_age_bucket(age_2018)
    if annot1_age_grp != age_2018_grp:
        return False
    return True
    
annot_join['annot_goodness'] = annot_join.apply(lambda row: check_for_annot_match(row), axis=1)
```

```python
annot_join['annot_goodness'].value_counts()
```

```python
good_dob = annot_join[annot_join['annot_goodness']==True]
```

```python
mismatch_age = len(good_dob[(good_dob['min_age'] > good_dob ['real_age']) | (good_dob['max_age'] < good_dob ['real_age'])])
tot = len(good_dob)
print("accuracy of age predication is " + str((tot-mismatch_age)/tot))
```

# Panel 5K random

```python
panel = pd.read_csv("data\\5k_panel_data_sample\\active_panelists_5k_sample.tsv", sep='\t')
panel.head(2)
```

```python
panel.shape
```

```python
import time

result_panel = []
err_cnt = 0 
cnt = 0 
new_window = True
for user in panel['panel_uid']:
    if new_window:
        #starting measuring time for rate limiter
        start = time.time()
        new_window = False
    else:
        # check if we passed the 15 min mark
        end = time.time()
        if (end - start) >= 900:
            new_window = True
        elif cnt == 900:
            # wait until new window for new limit of 900
            time.sleep(900 - (end - start))
            new_window = True
    try:
        error = False
        user_obj = api.GetUser(user)
    except:
        error = True
        err_cnt += 1
        print(user)
    if not error:
        #save for m3 as json
        with open("data\\twitter_cache\\input_" + str(user) + ".jsonl", "w") as text_file:
            text_file.write(str(user_obj))
        # prep json, this also downloads photo
        m3twitter.transform_jsonl(input_file="data\\twitter_cache\\input_" + str(user) + ".jsonl",output_file="data\\twitter_cache\\m3_input_" + str(user) + ".jsonl")
        result_panel.append(m3twitter.infer("data\\twitter_cache\\m3_input_" + str(user) + ".jsonl", output_format='dataframe'))                 
    cnt += 1
    if cnt%100 == 0:
        #save every 100
        result_panel_concat = pd.concat(result_panel)
        # save results
        result_panel_concat.to_csv("data\\m3_inference_on_panel.csv")
print("Number of fetch errors: " + str(err_cnt))        
```

```python
result_panel_concat = pd.concat(result_panel)
# save results
result_panel_concat.to_csv("data\\m3_inference_on_panel.csv")
```
